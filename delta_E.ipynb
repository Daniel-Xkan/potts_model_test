### import and initialize the paramters for position and amino acid for mutation


```python
import numpy as np

J = np.load('J.npy')
with open('in.nl43.reduce4.seq', 'r') as f:
    sequence = f.read().strip()

amino_acids = ['A', 'B', 'C', 'D']

position = 155
old_amino_acid = 'C'
new_amino_acid = 'D'

total_energy_change = 0
```

```python
J
```

```python
print(f"262: {sequence[262]}")
print(f"155: {sequence[155]}")
```

```python
#dictionary of J matrix 
J_dict = {}

row = 0
for pos1 in range(1, 263):
    for pos2 in range(pos1 + 1, 263):

        for i, aa1 in enumerate(['A', 'B', 'C', 'D']):
            for j, aa2 in enumerate(['A', 'B', 'C', 'D']):
                col = i*4+j
                

                J_dict[(pos1, pos2, aa1, aa2)] = J[row, col]
                J_dict[(pos2, pos1, aa2, aa1)] = J[row, col]
        print(f"Row {row}: J[{row}, {col}] = {J[row, col]} for positions ({pos1}, {pos2}) with amino acids ({aa1}, {aa2})")
        row += 1

print(f"Dictionary created with {len(J_dict)} entries")
```

sanity check: 34191 (262C2) *16 *2 = 1094112

```python
print( J_dict.get((0, 1, 'A', 'A'))) #shound be None
print( J_dict.get((1, 2, 'A', 'A')))
print( J_dict.get((1, 3, 'A', 'A')))
print( J_dict.get((261, 262, 'D', 'D')))
matching_keys = [key for key, value in J_dict.items() if abs(value - 0.022138428) < 0]
print(f"Keys with value 0.022138428: {matching_keys}")
```

```python
# Calculate energy for position 155 with original amino acid C
energy_155_C = 0
for other_pos in range(1, 263):
    if other_pos == 155:
        continue
    other_aa = sequence[other_pos - 1]
    energy_155_C += J_dict.get((155, other_pos, 'C', other_aa), 0)

# Calculate energy for position 155 with mutant amino acid D
energy_155_D = 0
for other_pos in range(1, 263):
    if other_pos == 155:
        continue
    other_aa = sequence[other_pos - 1]
    energy_155_D += J_dict.get((155, other_pos, 'D', other_aa), 0)

print(f"E(C) at 155: {energy_155_C}")
print(f"E(D) at 155: {energy_155_D}")
print(f"Delta E for C155D: {energy_155_D - energy_155_C}")
```
